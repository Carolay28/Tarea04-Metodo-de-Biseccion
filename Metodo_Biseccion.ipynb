## Ejercicio 1

Use el mÃ©todo de bisecciÃ³n para encontrar soluciones precisas dentro de $10^{-2}$ para $ð‘¥^3$ âˆ’ $7ð‘¥^2$ + 14ð‘¥ âˆ’ 6 = 0 en cada intervalo.

a. [0, 1]

```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

a = 0
b = 1
error = 1

while error > 1e-5:
    c = (a + b)/2
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c
    error = abs(b - a)

print("RaÃ­z aproximada en [0,1]:", (a+b)/2)
```

$x^2$ âˆ’ 1 = $e^{1-x^2}$.

## Ejercicio 1

Un abrevadero de longitud ð¿ tiene una secciÃ³n transversal en forma de semicÃ­rculo con radio ð‘Ÿ. (Consulte la figura adjunta.) Cuando se llena con agua hasta una distancia â„Ž a partir de la parte superior, el volumen ð‘‰ de agua es
ð‘‰ = ð¿ [0.5ðœ‹$r^2$ âˆ’ $r^2$ð‘Žð‘Ÿð‘ð‘ ð‘’ð‘›$(h/r)$ âˆ’ â„Ž$(ð‘Ÿ^2 - â„Ž^2)^{1/2}$].

Suponga que ð¿ = 10 ð‘ð‘š, ð‘Ÿ = 1 ð‘ð‘š y ð‘‰ = 12.4 $ð‘ð‘š^3$. Encuentre la profundidad del agua en el abrevadero
dentro de 0.01 ð‘ð‘š. 

```python
import math

L = 10
r = 1
V_deseado = 12.4
tolerancia = 0.01

def V(h):
    return L * (0.5*math.pi*r**2 - r**2*math.asin(h/r) - h*math.sqrt(r**2 - h**2))

a = 0
b = r
error = b - a

while error > tolerancia:
    c = (a + b)/2
    f_c = V(c) - V_deseado
    f_a = V(a) - V_deseado
    if f_a * f_c < 0:
        b = c
    else:
        a = c
    error = b - a

h_aprox = (a + b)/2
print("Profundidad aproximada del agua es h =", round(h_aprox, 2), "cm")
```

## Ejercicio 2

Un objeto que cae verticalmente a travÃ©s del aire estÃ¡ sujeto a una resistencia viscosa, asÃ­ como a la fuerza de gravedad. Suponga que un objeto con masa ð‘š cae desde una altura $ð‘ _0$ y que la altura del objeto despuÃ©s
de ð‘¡ segundos es:

$s(t) = s_0 - \frac{mg}{k}t + \frac{m^2g}{k^2} \left(1 - e^{-\frac{kt}{m}}\right)$,

donde ð‘” = 9.81 ð‘š/$s^2$ y ð‘˜ representa el coeficiente de la resistencia del aire en ð‘ð‘ /ð‘š. Suponga $ð‘ _0$ = 300 ð‘š,
ð‘š = 0.25 ð‘˜ð‘” y ð‘˜ = 0.1 ð‘ð‘ /ð‘š. Encuentre, dentro de 0.01 ð‘ ð‘’ð‘”ð‘¢ð‘›ð‘‘ð‘œð‘ , el tiempo que tarda un cuarto de kg en
golpear el piso. 


```python
import math

s0 = 300      
m = 0.25      
k = 0.1       
g = 9.81      
tolerancia = 0.01

def s(t):
    return s0 - (m*g/k)*t + (m**2 * g / k**2)*(1 - math.exp(-k*t/m))

a = 0
b = 1000  
error = b - a

while error > tolerancia:
    c = (a + b)/2
    f_c = s(c)
    f_a = s(a)
    if f_a * f_c < 0:
        b = c
    else:
        a = c
    error = b - a

t_aprox = (a + b)/2
print("Tiempo aproximado de caÃ­da:", round(t_aprox, 2), "segundos")
```

## Ejercicio 1

Use el teorema 2.1 para encontrar una cota para el nÃºmero de iteraciones necesarias para lograr una aproximaciÃ³n con precisiÃ³n de $10^{-4}$ para la soluciÃ³n de $x^3$ âˆ’ ð‘¥ âˆ’ 1 = 0 que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximaciÃ³n para la raÃ­z con este grado de precisiÃ³n. 


```python
def f(x):
    return x**5 - x - 1

a = 1
b = 2
n_iter = 20

for _ in range(n_iter):
    c = (a + b)/2
    if f(a)*f(c) < 0:
        b = c
    else:
        a = c

raiz_aprox = (a + b)/2
print("AproximaciÃ³n de la raÃ­z:", round(raiz_aprox, 6))
```

## Ejercicio 2

La funciÃ³n definida por ð‘“(ð‘¥) = sin ðœ‹ ð‘¥ tiene ceros en cada entero. Muestre cuando âˆ’1 < ð‘Ž < 0 y 2 < ð‘ < 3, el mÃ©todo de bisecciÃ³n converge a:

a. 0, ð‘ ð‘– ð‘Ž + ð‘ < 2 

```python
import math

def f(x):
    return math.sin(math.pi * x)

a = -0.5
b = 2.0  
tolerancia = 0.00001

while (b - a) > tolerancia:
    c = (a + b) / 2
    if f(c) == 0:
        break
    elif f(a) * f(c) < 0:
        b = c
    else:
        a = c

print("La raÃ­z aproximada es:", c)
```
